## 1.2.Transformación (Transform)

```python
#Resumen de la información del DataFrame
df.info()
```

```python
columnas_object = ['anfitrión_es_superanfitrión', 'barrio', 'código_postal', 'tipo_propiedad', 'tipo_habitación', 'servicios', 'precio', 'tiene_disponibilidad', 'fecha_ingreso', 'hora_ingreso', 'fecha_egreso', 'hora_egreso']

#seleccionar las columnas y mostrar su información
display(df[columnas_object].info())

#Verificamos que el dato "object" es realmente un "str"

#para verificar el tipo de dato de un elemento en cada una:
for col in columnas_object:
    if col in df.columns:
        print(f"Tipo de dato de '{col}': {type(df[col].iloc[0])}")
    else:
        print(f"La columna '{col}' no se encuentra en el DataFrame.")
```

```python
# Crear una columnas de fecha y hora combinadas y convertirlas al tipo datatime
df['fecha_hora_ingreso'] = pd.to_datetime(df['fecha_ingreso'] + ' ' + df['hora_ingreso'], format='%Y-%m-%d %H:%M:%S')
df['fecha_hora_egreso'] = pd.to_datetime(df['fecha_egreso'] + ' ' + df['hora_egreso'], format='%Y-%m-%d %H:%M:%S')
```

```python
print(df.columns)
```

```python
# Eliminar las columnas 'fecha' y 'hora' separadas
df.drop(columns=['fecha_ingreso','hora_ingreso', 'fecha_egreso', 'hora_egreso'],inplace=True)
```

```python
df.info()
```

```python
display(df.head())
```

```python
# Limpiar la columna 'precio' eliminando el símbolo de dólar y las comas, si existen
df['precio_limpio'] = df['precio'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Convertir la columna limpia a float64
df['precio_float'] = pd.to_numeric(df['precio_limpio'], errors='coerce')

# Eliminar la columna intermedia 'precio_limpio' si no es necesaria
df.drop(columns=['precio_limpio'], inplace=True, errors='ignore')

# Mostrar las primeras filas y la información del DataFrame para verificar la conversión
display(df[['precio', 'precio_float']].head())
display(df.info())
```

```python
# Eliminar la columna original 'precio'
if 'precio' in df.columns:
    df.drop(columns=['precio'], inplace=True)
    print("Columna 'precio' eliminada exitosamente.")
else:
    print("La columna 'precio' no se encuentra en el DataFrame.")

display(df.head())
display(df.info())
```

```python
# Establecer la opción de visualización para mostrar floats con 2 decimales
pd.options.display.float_format = '{:.2f}'.format

# Mostrar las primeras filas para verificar el formato
display(df[['precio_float']].head())
```

# 3-PREGUNTAS DE NEGOCIO

```python
# Pregunta 1: Precio promedio por barrio
precio_promedio_por_barrio = df.groupby('barrio')['precio_float'].mean().sort_values(ascending=False)
print("Precio promedio de alquiler por barrio:")
display(precio_promedio_por_barrio)

# Conclusión  1:
print("\nConclusión 1: El análisis muestra que los precios promedio de alquiler varían significativamente entre barrios, siendo algunos considerablemente más caros que otros.")
```

```python
# Pregunta 2: Capacidad promedio por tipo de propiedad
capacidad_promedio_por_tipo_propiedad = df.groupby('tipo_propiedad')['capacidad'].mean().sort_values(ascending=False)
print("\nCapacidad promedio por tipo de propiedad:")
display(capacidad_promedio_por_tipo_propiedad)

# Conclusión  2:
print("\nConclusión 2: La capacidad promedio varía según el tipo de propiedad, lo cual es esperable ya que algunos tipos (como las casas) suelen ser más grandes que otros (como los apartamentos).")
```

```python
# Pregunta 3: Disponibilidad a corto y largo plazo
disponibilidad_promedio = df[['disponibilidad_30', 'disponibilidad_60', 'disponibilidad_90', 'disponibilidad_365']].mean()
print("\nDisponibilidad promedio:")
display(disponibilidad_promedio)

# Conclusión  3:
print("\nConclusión 3: La disponibilidad promedio aumenta a medida que se considera un horizonte de tiempo más amplio, lo que indica que hay más propiedades disponibles para alquiler a largo plazo que a corto plazo.")
```
